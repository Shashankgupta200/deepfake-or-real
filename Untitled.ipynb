```python
import os
import shutil
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# Paths and parameters
base_dir = "./data"  # Correct path to your dataset
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'val')
test_dir = os.path.join(base_dir, 'test')

IMG_HEIGHT = 256
IMG_WIDTH = 256
BATCH_SIZE = 64
EPOCHS = 21
PATIENCE = 5  # Early stopping patience

# Create directories for train, val, and test splits
def create_split_directories():
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    for category in ['real', 'fake']:
        os.makedirs(os.path.join(train_dir, category), exist_ok=True)
        os.makedirs(os.path.join(val_dir, category), exist_ok=True)
        os.makedirs(os.path.join(test_dir, category), exist_ok=True)

# Split images into train, validation, and test sets
def split_data():
    # List the images in each class folder
    for category in ['real', 'fake']:
        category_path = os.path.join(base_dir, category)
        files = [os.path.join(category_path, f) for f in os.listdir(category_path)]


        # Split into train (70%), validation (15%), and test (15%)
        train_files, temp_files = train_test_split(files, test_size=0.3, random_state=42)
        val_files, test_files = train_test_split(temp_files, test_size=0.5, random_state=42)

        # Move files to respective directories
        for file in train_files:
            shutil.move(file, os.path.join(train_dir, category, os.path.basename(file)))
        for file in val_files:
            shutil.move(file, os.path.join(val_dir, category, os.path.basename(file)))
        for file in test_files:
            shutil.move(file, os.path.join(test_dir, category, os.path.basename(file)))

# Data Augmentation with validation_split=0.2
datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest",
)

# Create splits and move files
create_split_directories()
split_data()

# Train (70% of the data)
train_data = datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=True,
)

# Validation (15% of the data)
val_data = datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=True,
)

# Test data (15% of the data)
test_datagen = ImageDataGenerator(rescale=1.0 / 255)  # No validation split for test data
test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=True,
)

# Print dataset sizes
print(f"Number of training images: {train_data.samples}")
print(f"Number of validation images: {val_data.samples}")
print(f"Number of test images: {test_data.samples}")

# DenseNet Model
base_model = DenseNet121(
    weights="imagenet", include_top=False, input_shape=(IMG_HEIGHT, IMG_WIDTH, 3)
)

# Freeze all layers initially
base_model.trainable = False

# Unfreeze the last 15 layers of the DenseNet model
for layer in base_model.layers[-15:]:
    layer.trainable = True

# Build the model
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(128, activation="relu", kernel_regularizer=l2(0.01)),
    Dropout(0.5),
    Dense(1, activation="sigmoid", kernel_regularizer=l2(0.01)),
])

# Compile the model
model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"],
)

# Early stopping callback
early_stopping = EarlyStopping(
    monitor="val_loss", patience=PATIENCE, restore_best_weights=True
)

# Train the model
history = model.fit(
    train_data,
    epochs=EPOCHS,
    validation_data=val_data,
    callbacks=[early_stopping],
)

# Save the trained model
model.save("deepfake_detector_model.h5")
print("Model saved successfully.")

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(test_data)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Plot training history
def plot_history(history):
    acc = history.history["accuracy"]
    val_acc = history.history["val_accuracy"]
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]

    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, "b", label="Training Accuracy")
    plt.plot(epochs, val_acc, "r", label="Validation Accuracy")
    plt.title("Training and Validation Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, "b", label="Training Loss")
    plt.plot(epochs, val_loss, "r", label="Validation Loss")
    plt.title("Training and Validation Loss")
    plt.legend()

    plt.show()

plot_history(history)
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

# Load the trained model
model = load_model("deepfake_detector_model.h5")

IMG_HEIGHT = 256
IMG_WIDTH = 256

# Function to preprocess and predict the image
def predict_image(image_path, model):
    # Load the image with target size to match the model input
    img = image.load_img(image_path, target_size=(IMG_HEIGHT, IMG_WIDTH))
    
    # Convert the image to a numpy array and rescale pixel values (same as during training)
    img_array = image.img_to_array(img) / 255.0  # Rescaling pixel values between 0 and 1
    
    # Add a batch dimension as the model expects a batch of images
    img_array = np.expand_dims(img_array, axis=0)
    
    # Make the prediction
    prediction = model.predict(img_array)
    
    # Output the result, flipped categories for 'Real' and 'Deepfake'
    if prediction[0] > 0.5:
        return "Real"  # Prediction for real image
    else:
        return "Deepfake"  # Prediction for deepfake image

# Test the prediction function
image_path = "/Users/shashankgupta/Downloads/Dataset/Test/Fake/fake_1.jpg"  # Replace with the path to the image you want to test
result = predict_image(image_path, model)

print(f"Prediction: {result}")

# Optionally, show the image
img = image.load_img(image_path)
plt.imshow(img)
plt.title(f"Prediction: {result}")
plt.show()
```
